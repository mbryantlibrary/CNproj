```python
import numpy as np
```

```python
def choose_indices():
        a = np.random.randint(100)
        b = 0
        while(b==0):
            b = np.random.randint(2 * 5) - 5
        b = a + b
        if(b < 0):
            b = 100 + b
        elif(b >= 100):
            b -= 100
        return (a,b)       
```

```python
a,b = 0,0
count = 0
while(a != 100 and b != 100 and count < 10000):
    a,b = choose_indices()
    print(a,b)
    count += 1
```
