```python
import pandas as pd
def getBestGenotypeFromFile(file):
    alldata = pd.read_csv(file)
    nCol = alldata.shape[1]
    colList = ['Fitness']
    colList.extend(["g"+str(i) for i in range(0,nCol-1)])
    alldata.columns = colList
    return Genotype(alldata.sort('Fitness',ascending=False).T.iloc[1:,0].values)

geno = getBestGenotypeFromFile('\exp\22.04.15_2.csv')
netp = NetParams(geno)
```

```python
def doRun(trialtype):
    netp.reset_weights()
    #weights are initialised to random at the start of the run
    #but are persistent throughout the rest of the run to allow plasticity
    print("Starting run of trial type %d" % trialtype)
    print("Initial weights %s" % str(netp.weights))
    print("5 unevaluated runs:")
    for i in range(0,5):
        #these trials are not evaluated, just for plasticity
        f = Trial(netp,trialtype).run()
        print("\t%d: %f" % (i,f))
        print("\t w: %s" % str(netp.weights))
    sumFit = 0
    print("3 evaluated runs:")
    for i in range(0,3):
        #evaluate three trials for fitness
        t = Trial(netp,trialtype)
        sumFit += t.run()
        print("\t%d: %f" % (i,f))
        print("\t w: %s" % str(netp.weights))
    return sumFit/3 #average fitness

trials = []
sumFit = 0
sumFit += doRun(TrialType.LIGHTA)
sumFit += doRun(TrialType.LIGHTB)
sumFit += doRun(TrialType.BOTH_BLINKB)
sumFit += doRun(TrialType.BOTH_BLINKA)
```

```python
dists = []
fitHs = []
fitPs = []
fitDs = []
weights = []

for t in range(0,60):
    trial = Trial(netp,TrialType.BOTH_BLINKB)
    trial.run()
    dists.append(trial.getDists())
    fitHs.append(trial.fitH)
    fitPs.append(trial.fitP)
    fitDs.append(trial.fitD)
    print("trial %d/%d:fH=%f,fP=%f,fD=%f,dist=%s" % (t,60,fitHs[-1],fitPs[-1],fitDs[-1],str(dists[-1])))
    weights.append(trial.controller.nn.netp.weights)
    print(weights[-1])
```

```python
plt.plot(range(0,len(dists)),dists)
plt.figure()
plt.plot(range(0,len(fitHs)),fitHs)
plt.figure()
plt.plot(range(0,len(fitPs)),fitPs)
plt.figure()
plt.plot(range(0,len(fitDs)),fitDs)
plt.figure()
plt.plot(range(0,len(fitDs)),fitDs)

```
